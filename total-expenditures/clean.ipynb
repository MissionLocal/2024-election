```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.lower()
df.amount = df.amount.str.replace('$', '')
df.amount = df.amount.str.replace(',', '')
df.amount = df.amount.str.replace('(', '-')
df.amount = df.amount.str.replace(')', '')
df.amount = df.amount.astype(float)
```

```python
df.tx_date = pd.to_datetime(df.tx_date)
```

```python
donors = df.name.value_counts().reset_index()
```

```python
donors
```

```python
donors.to_csv('donors.csv')
```

```python
select = ["GROWSF VOTER GUIDE", "Independent Expenditure by Coalition to Grow San Francisco - Grow SF PAC", "COALITION TO GROW SAN FRANCISCO - GROW SF", "Independent Expenditure by Families for a Vibrant SF Sponsored by Abundance Network", "ABUNDANCE NETWORK", "Independent Expenditure by FORWARD ACTION SF, SUPPORTING LONDON BREED FOR MAYOR 2024, SPONSORED BY ABUNDANCE NETWORK", "Independent Expenditure by TogetherSF Action", "TOGETHERSF ACTION", "TOGETHER SF", "TOGETHERSF"]
```

```python
filtered_df = df[df["name"].isin(select)]
```

```python
filtered_df.amount.sum()
```

```python
filtered_df.groupby('name').amount.sum()
```

```python
filtered_df.groupby('candidate_or_measure').amount.sum().sort_values(ascending=False)
```

```python
clean_df = filtered_df[['name', 'amount', 'candidate_or_measure', 'tx_date', 'committee_name', "position"]]
```

```python
# show em the rows with NaNs
clean_df[clean_df.isnull().any(axis=1)]
```

```python
# remove rows with NaNs in the amount column

clean_df = clean_df.dropna(subset=['amount'])
```

```python
# remove rows where the amount is 0 

clean_df = clean_df[clean_df.amount != 0]
```

```python
clean_df[clean_df.isnull().any(axis=1)]
```

```python
# drop any rows with NaNs

clean_df = clean_df.dropna()
```

```python
clean_df
```

```python
# only keep the rows where the date is from september 2024 onwards

clean_df = clean_df[clean_df.tx_date >= '2024-09-01']
```

```python
clean_df.name.value_counts()
```

```python
tags = {
    "Independent Expenditure by Coalition to Grow San Francisco - Grow SF PAC": "GrowSF",
    "COALITION TO GROW SAN FRANCISCO - GROW SF": "GrowSF",
    "Independent Expenditure by TogetherSF Action": "TogetherSF",
    "Independent Expenditure by Families for a Vibrant SF Sponsored by Abundance Network": "Abundance Network",
    "TOGETHERSF ACTION": "TogetherSF",
    "GROWSF VOTER GUIDE": "GrowSF",
    "ABUNDANCE NETWORK": "Abundance Network",
    "COALITION TO GROW SAN FRANCISCO - GROW SF": "GrowSF",
    "TOGETHER SF": "TogetherSF",
    "TOGETHERSF": "TogetherSF",
    "Independent Expenditure by FORWARD ACTION SF, SUPPORTING LONDON BREED FOR MAYOR 2024, SPONSORED BY ABUNDANCE NETWORK": "Abundance Network"
}
```

```python
clean_df['tag'] = clean_df.name.map(tags)
```

```python
clean_df.to_csv('clean.csv', index=False)
```
